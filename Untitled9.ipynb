```python
# Task 2

print('Table  \n\n')
print('Val'+"                "+'Event\n')
table=[]
for i in range(len(triggers['NAME'])):
  if triggers['NAME'][i]=='IGN_CYL':
    k=triggers['VAL'][i]
    if k=="ON":
      l=["ON","ignitionOn"]
    else:
      l=["OFF",'ignitionOff']
    table.append(l)
for item in table:
  print(item[0]+"                "+item[1])
```
